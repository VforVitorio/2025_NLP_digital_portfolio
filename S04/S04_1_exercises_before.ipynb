# Tokenization

Split a text into words.

```python
# 1 Exercise: Split the file cantares_gallegos.txt into words. 

with open("cantares_gallegos.txt", "r") as cantares_f:
    words = cantares_f.read().split()
    print(words)

```

```python
# 2 Exercise: Count the number of ocurrences of each word in the file cantares_gallegos.txt.

def repeated_words(words):
    rep_words = {}
    for word in words:
        if word in rep_words:
            rep_words[word] +=1
        else:
            rep_words[word] = 1
    return rep_words

print(repeated_words(words))
    
```

```python
# 3 Exercise: Order the words by number of ocurrences in the file cantares_gallegos.txt and print the 20 most common words.


# source of solution: https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value
new_words = repeated_words(words)

sorted_occurences = dict(sorted(new_words.items(), key=lambda item: item[1], reverse=True))

print(sorted_occurences)

```
